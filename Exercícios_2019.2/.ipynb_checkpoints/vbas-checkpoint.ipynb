```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn import svm
```

```python
tips = sns.load_dataset('tips')
X = tips.drop('tip', axis = 1)
y = tips['tip']
tips.head()
```

```python
lb = LabelEncoder()
X['sex'] = lb.fit_transform(X['sex'])
X['smoker'] = lb.fit_transform(X['smoker'])
X['day'] = lb.fit_transform(X['day'])
X['time'] = lb.fit_transform(X['time'])
X.head()
```

```python
plt.figure(dpi=100)
sns.distplot(tips['tip'])
plt.title('Tips Distribuition')
plt.show()
```

```python
tips['sex'] = X['sex']
tips['smoker'] = X['smoker']
tips['day'] = X['day']
tips['time'] = X['time']

f, (ax1,ax2) = plt.subplots(1,2, dpi=180)
sns.heatmap(tips.corr(), ax=ax1)
ax1.set_title('All')
sns.heatmap(tips.drop(['sex','smoker','day','time'], axis=1).corr(), ax=ax2)
ax2.set_title('Numerical')

plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
X.head()
```

```python
X_train = np.asarray(X_train)
y_train = np.asarray(y_train)
X_test = np.asarray(X_test)
y_test = np.asarray(y_test)

clf = svm.SVR(C=100)
clf.fit(X_train, y_train)
train_pred = clf.predict(X_train)
test_pred = clf.predict(X_test)

f, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(16,4))
ax1.plot(y_train, label='train data')
ax1.plot(train_pred, 'r--', label='train fitted data',alpha=.5)
ax1.set_title('train fitted data')
ax1.legend()

ax2.plot(y_test, label='test data')
ax2.plot(test_pred, 'r--', label='predicted')
ax2.set_title('predicted test data')
ax2.legend()

ax3.plot(y_test, label='test data')
ax3.scatter(range(0, len(test_pred)), test_pred, marker='o', color='red', label='predicted scatter', s=10)
ax3.set_title('predicted test data')
ax3.legend()

sns.distplot(y_test-test_pred, ax=ax4)
ax4.set_title('Distribution plot of the difference\nbetween the real and predicted tips value')

plt.tight_layout()
plt.show()
```
